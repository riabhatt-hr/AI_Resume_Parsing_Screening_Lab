```python
# This cell imports the libraries required in this lab.
# pandas is used for data handling.
# re is used for pattern matching and simple text extraction.
# sklearn is used for TF-IDF and cosine similarity.

import pandas as pd
import re

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Display settings to make output easier to read.
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1000)

print("Libraries imported successfully.")
```

This step helps us inspect:

## Step 4: Parse Skills from Resume Text
Now we simulate resume parsing.
We extract the skills section from the unstructured text.

```python
# This cell extracts skills from each resume using simple text parsing.
# It looks for the text between "skills in" and the next period.

def extract_skills(text):
    match = re.search(r"skills in (.*?)[.] He has|skills in (.*?)[.] She has", text, re.IGNORECASE)
    if match:
        extracted = match.group(1) if match.group(1) else match.group(2)
        return extracted
    return ""

df["skills_extracted"] = df["resume_text"].apply(extract_skills)

df[["candidate_id", "skills_extracted"]].head()
```

## Step 5: Parse Experience from Resume Text
In this step, we extract the years of experience from each resume.

```python
# This cell extracts numeric years of experience from the resume text.
# It searches for patterns such as "3 years of experience".

def extract_experience(text):
    match = re.search(r"(\d+)\s+years of experience", text, re.IGNORECASE)
    if match:
        return int(match.group(1))
    return 0

df["experience_extracted"] = df["resume_text"].apply(extract_experience)

df[["candidate_id", "experience_extracted"]].head()
```

## Step 6: Parse Education from Resume Text
Now we extract educational qualification from the resume text.

```python
# This cell extracts education information from the resume text.
# It searches for the value written after "Education:".

def extract_education(text):
    match = re.search(r"Education:\s*([A-Za-z. ]+)", text)
    if match:
        return match.group(1).strip()
    return ""

df["education_extracted"] = df["resume_text"].apply(extract_education)

df[["candidate_id", "education_extracted"]].head()
```

## Step 7: Parse Certifications from Resume Text
This step extracts the number of certifications from each resume.

```python
# This cell extracts the certification count from each resume.
# It searches for patterns like "Certifications: 2"

def extract_certifications(text):
    match = re.search(r"Certifications:\s*(\d+)", text, re.IGNORECASE)
    if match:
        return int(match.group(1))
    return 0

df["certifications_extracted"] = df["resume_text"].apply(extract_certifications)

df[["candidate_id", "certifications_extracted"]].head()
```

## Step 8: View Structured Parsed Resume Data
Now we see how unstructured resume text has been transformed into structured candidate information.

```python
# This cell displays the parsed structured data.

parsed_df = df[[
    "candidate_id",
    "resume_text",
    "skills_extracted",
    "experience_extracted",
    "education_extracted",
    "certifications_extracted"
]]

parsed_df.head(10)
```

## Step 9: Define a Job Description
This job description acts as the target profile for matching.

```python
# This cell defines the job description.
# The system will compare candidate resumes with this text.

job_description = "python sql machine learning nlp communication recruitment analytics power bi"

print("Job Description:")
print(job_description)
```

## Step 10: Perform ATS-Style Keyword Matching
An ATS often checks whether candidate resumes contain the required keywords.
We now simulate this process.

```python
# This cell defines required keywords from the job description.
# It then counts how many of these keywords appear in the extracted skills.

required_keywords = ["python", "sql", "machine learning", "nlp", "communication", "recruitment", "analytics", "power bi"]

def keyword_match_count(skill_text):
    skill_text_lower = skill_text.lower()
    count = 0
    for keyword in required_keywords:
        if keyword in skill_text_lower:
            count += 1
    return count

df["keyword_match_count"] = df["skills_extracted"].apply(keyword_match_count)

df[["candidate_id", "skills_extracted", "keyword_match_count"]].head(10)
```

## Step 11: Convert Resume Skills and Job Description into TF-IDF Vectors
Beyond exact keyword matching, AI can compare text similarity.
This helps capture broader relevance.

```python
# This cell converts the job description and extracted skills into TF-IDF vectors.

documents = [job_description] + df["skills_extracted"].tolist()

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

print("TF-IDF matrix shape:", tfidf_matrix.shape)
```

## Step 12: Calculate Cosine Similarity
This step measures how similar each candidate's skills are to the job description.

```python
# This cell calculates cosine similarity between the job description
# and each candidate's extracted skills.

job_vector = tfidf_matrix[0:1]
candidate_vectors = tfidf_matrix[1:]

similarity_scores = cosine_similarity(job_vector, candidate_vectors).flatten()

df["similarity_score"] = similarity_scores

df[["candidate_id", "skills_extracted", "similarity_score"]].head(10)
```

## Step 13: Normalize Experience and Certifications
To combine different candidate attributes fairly, we scale them to the 0–1 range.

```python
# This cell normalizes extracted experience and certification values.

df["experience_score"] = df["experience_extracted"] / df["experience_extracted"].max()
df["certification_score"] = df["certifications_extracted"] / df["certifications_extracted"].max()

df[["candidate_id", "experience_extracted", "experience_score", "certifications_extracted", "certification_score"]].head()
```

## Step 14: Convert Education into a Numerical Score
Education is categorical, so we assign simple scores for demonstration.

```python
# This cell maps extracted education values to numerical scores.
# This is only for teaching purposes.

education_map = {
    "B.Com": 0.50,
    "BBA": 0.55,
    "BCA": 0.55,
    "B.Sc": 0.60,
    "B.Tech": 0.70,
    "MBA": 0.75,
    "MBA HR": 0.80,
    "MCA": 0.80,
    "M.Sc": 0.80,
    "M.Tech": 0.90,
    "M.Stat": 0.90
}

df["education_score"] = df["education_extracted"].map(education_map)

df[["candidate_id", "education_extracted", "education_score"]].head(10)
```

## Step 15: Create Final Resume Matching Score
Now we combine:
- keyword matching
- text similarity
- experience
- education
- certifications

This gives a final candidate relevance score.

```python
# This cell creates a final weighted score.
# We first normalize the keyword match count.
# Then we combine all scores using weights.

df["keyword_score"] = df["keyword_match_count"] / len(required_keywords)

df["final_score"] = (
    0.30 * df["keyword_score"] +
    0.30 * df["similarity_score"] +
    0.20 * df["experience_score"] +
    0.10 * df["education_score"] +
    0.10 * df["certification_score"]
)

df[[
    "candidate_id",
    "keyword_score",
    "similarity_score",
    "experience_score",
    "education_score",
    "certification_score",
    "final_score"
]].head(10)
```

## Step 16: Rank Candidates
Now we sort candidates from highest final score to lowest final score.

```python
# This cell ranks candidates based on final score.

ranked_df = df.sort_values(by="final_score", ascending=False).reset_index(drop=True)

ranked_df[[
    "candidate_id",
    "skills_extracted",
    "experience_extracted",
    "education_extracted",
    "certifications_extracted",
    "final_score"
]]
```

## Step 17: Identify Top 5 Best-Fit Candidates
These are the strongest candidates according to the ATS-style AI screening logic.

```python
# This cell extracts the top 5 candidates.

top_5 = ranked_df.head(5)

top_5[[
    "candidate_id",
    "skills_extracted",
    "experience_extracted",
    "education_extracted",
    "certifications_extracted",
    "final_score"
]]
```

## Step 18: Add Recommendation Labels
This makes the output more interpretable for HR.

```python
# This cell converts numeric final scores into recommendation labels.

def label_candidate(score):
    if score >= 0.65:
        return "Strong Fit"
    elif score >= 0.50:
        return "Moderate Fit"
    else:
        return "Low Fit"

ranked_df["recommendation"] = ranked_df["final_score"].apply(label_candidate)

ranked_df[[
    "candidate_id",
    "final_score",
    "recommendation"
]]
```

## Step 19: Final Structured Resume Screening Output
This output represents the parsed and ranked candidate dataset.

```python
# This cell creates the final output table.

final_output = ranked_df[[
    "candidate_id",
    "resume_text",
    "skills_extracted",
    "experience_extracted",
    "education_extracted",
    "certifications_extracted",
    "keyword_match_count",
    "keyword_score",
    "similarity_score",
    "experience_score",
    "education_score",
    "certification_score",
    "final_score",
    "recommendation"
]]

final_output
```
